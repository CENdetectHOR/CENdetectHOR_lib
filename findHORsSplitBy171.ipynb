```python
import numpy as np
import matplotlib.pyplot as plt
from distanceMatrix import build_distance_matrix, matrix_closure, distance_values, clusterings_with_hors
```

```python
fasta_file = open("./data/chr11_HSA_simple.fasta")
seqStr = fasta_file.read()
fasta_file.close()
```

```python
code_map = {'A': 0, 'C': 1, 'G': 2, 'T': 3}
code_seq = np.array([code_map[b] for b in seqStr])
train_seq = code_seq.reshape(1, -1)
```

```python
monomer_size = 171
dna_seqs = [code_seq[sq_index * monomer_size : (sq_index + 1) * monomer_size] for sq_index in range(len(code_seq) // monomer_size)]
```

```python
dna_seq_dists = build_distance_matrix(dna_seqs)
```

```python
plt.hist(distance_values(dna_seq_dists), bins=60)
```

```python
plt.matshow(dna_seq_dists)
```

```python
plt.matshow(matrix_closure(dna_seq_dists<=35))
```

```python
clusterings = clusterings_with_hors(dna_seq_dists)
```
