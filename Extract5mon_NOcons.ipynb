```python
import numpy as np
import re
import random
```

```python
file="./chr11_HSA.fasta"
out="./output/extract5mon/"
mon=171
```

```python
seq=[]
with open(file, "r") as f:
    for l,line in enumerate(f):
        if l==0:
            line=line.strip()
            line=re.split(':|-|>',line)
            chrom=line[1]
            start=line[2]
            end=line[3]

        else:
            seq.append(line.strip())
seq=[x for y in seq for x in y]
len_seq=len(seq)
Nmax_mons=len_seq/mon
Nmin_mons=Nmax_mons-(Nmax_mons/100*30)
min_mon=mon-5
max_mon=mon+5
```

```python
s2c_list=[]
for p,pos in enumerate(seq):
    endpos=int(p)+6
    s2c_2app=seq[p:endpos]
    s2c_list.append("".join(s2c_2app))
```

```python
for index in s2c_list:
    start_list=[]
    mon_list=[]
    for b,base in enumerate(seq):
        checkend=int(b)+6
        start2check=seq[int(b):int(checkend)]
        s2c="".join(start2check)
        if s2c==index:
            start_list.append(b)
    for t,temp in enumerate(start_list):
        if t==0:
            prev=int(temp)
        else:
            mon2app=seq[int(prev):int(temp)]
            mon_list.append(mon2app)
            prev=temp
    fin_mon_list=[x for x in mon_list if len(x)>min_mon and len(x)<max_mon]
    if len(fin_mon_list)>Nmin_mons:
        break
    else:
        continue
```

```python
fin_mon_list=[x for x in fin_mon_list if len(x)==mon]
fin_mon_list=["".join(m) for m in fin_mon_list]
random_seq=random.sample(fin_mon_list, 5)
with open(out+chrom+"_5mons.fasta", "w") as o:
    for r,ran in enumerate(random_seq):
        o.write(">mon"+str(r+1)+"\n")
        o.write(ran+"\n")

```
